## Analysis 1: Average Final Grade (G3)

The `G3` column represents the final grade obtained by each student. In this step, we calculate the average final grade to understand the overall academic performance of the students in the dataset.

```python
average_grade = df["G3"].mean()
print("Average Final Grade (G3):", round(average_grade, 2))
```

### Observation

The average final grade (G3) of the students is **10.42**. This value represents the overall academic performance of all students in the dataset.

## Analysis 2: Students Scoring Above 15

In this step, we identify students who scored more than 15 in the final grade (G3). This helps determine the number of high-performing students in the dataset.

```python
students_above_15 = df[df["G3"] > 15].shape[0]
print("Number of students scoring above 15:", students_above_15)
```

### Observation

A total of **40** students scored above 15 in the final examination. These students can be considered high academic performers based on their final grades.

## Analysis 3: Relationship Between Study Time and Final Grade

This analysis examines whether there is a relationship between the amount of time students spend studying (`studytime`) and their final grades (`G3`). The Pearson correlation coefficient is used to measure the strength and direction of this relationship.

```python
correlation = df["studytime"].corr(df["G3"])
print("Correlation between Study Time and Final Grade:", round(correlation, 3))
```

### Observation

The correlation coefficient between study time and final grade is **0.098**.

This indicates a **very weak positive relationship** between study time and students' final grades. While students who study more tend to achieve slightly higher grades, the relationship is weak. This suggests that study time alone does not strongly influence academic performance, and other factors such as attendance, previous grades, family support, and motivation may also play significant roles.

## Analysis 4: Average Final Grade by Gender

This analysis compares the average final grades of male and female students to determine which group performed better on average.

```python
gender_average = df.groupby("sex")["G3"].mean().round(2)
print(gender_average)
```

### Observation

The average final grade was calculated separately for male and female students.

- **Female students:** 9.97
- **Male students:** 10.91

Based on the analysis, **male students performed slightly better on average** than female students in this dataset, with an average final grade of **10.91** compared to **9.97** for female students.

# Visualization 1: Distribution of Final Grades

A histogram is used to display the distribution of numerical data. In this analysis, it shows how the final grades (G3) are distributed among the students.

```python
plt.figure(figsize=(8,5))
plt.hist(df["G3"], bins=10, edgecolor="black")
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")

plt.show()
```

### Observation

The histogram shows the distribution of students' final grades (G3). Most students scored between **9 and 15**, with the highest concentration of grades around **10 to 12**. Only a small number of students achieved very high grades (above 17) or very low grades (below 5).

The graph also shows that several students received a grade of **0**, which may indicate failure or absence in the final examination. Overall, the distribution suggests that the majority of students achieved average academic performance, while relatively few students were either top performers or low performers.

# Visualization 2: Study Time vs Final Grade

A scatter plot is used to examine the relationship between two numerical variables. Here, it helps visualize whether students who spend more time studying tend to achieve higher final grades.

```python
plt.figure(figsize=(8,5))
plt.scatter(df["studytime"], df["G3"])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")

plt.show()
```

### Observation

The scatter plot illustrates the relationship between students' study time and their final grades (G3). Since the `studytime` variable is divided into four categories (1 to 4), the data points appear as four distinct vertical groups.

Students in every study time category achieved a wide range of final grades, from very low to very high. Although the highest study time category (4) includes some of the highest grades, similar grades are also observed in the lower study time categories.

This pattern indicates that there is only a **very weak positive relationship** between study time and final grade, which is supported by the calculated correlation coefficient of **0.098**. Therefore, study time alone does not strongly determine a student's academic performance, and other factors are likely to influence the final grades.

# Visualization 3: Average Final Grade by Gender

A bar chart is used to compare values across different categories. This chart compares the average final grades of male and female students.

```python
gender_average = df.groupby("sex")["G3"].mean()
plt.figure(figsize=(6,5))
plt.bar(gender_average.index, gender_average.values)
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")

plt.show()
```

### Observation

The bar chart compares the average final grades of male and female students. The analysis shows that male students achieved an average final grade of approximately **10.91**, while female students achieved an average of approximately **9.97**.

Although male students performed slightly better on average, the difference between the two groups is relatively small. This suggests that the overall academic performance of male and female students is fairly similar.
